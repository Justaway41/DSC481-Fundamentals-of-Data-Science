## 2. Write a function that returns both the minimum and maximum of a list.

```python
def min_max(lst):
    return min(lst), max(lst)

min_max([3, 1, 7, 2])
```

## 3. Write a recursive function to compute the factorial of a number.

```python
def factorial(n):
    if n <= 1:
        return 1
    return n * factorial(n-1)

factorial(5)
```

## 11. Given a list with duplicates, return a list with unique elements, preserving order.

```python
def dedupe(seq):
    seen = set()
    out = []
    for x in seq:
        if x not in seen:
            seen.add(x)
            out.append(x)
    return out

dedupe([1,2,2,3,1,4])
```

## 12. Use a dictionary comprehension to invert a dictionary (swap keys and values).

```python
d = {"a": 1, "b": 2, "c": 3}
inverted = {v: k for k, v in d.items()}
inverted
```
