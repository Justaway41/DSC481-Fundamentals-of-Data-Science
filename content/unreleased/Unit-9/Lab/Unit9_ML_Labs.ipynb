This notebook contains lab exercises for Unit IX: Introduction to Machine Learning with Python. You will practice regression, classification, and clustering using Scikit-learn.

```python
# Import Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.cluster import KMeans
```

## Simple Regression: Predicting Marks from Study Hours

Create a small dataset with study hours and marks. Fit a regression model and predict marks for new study hours. Visualize the regression line.

```python
# Create dataset
marks = [60, 65, 70, 75, 80]
study_hours = [2, 3, 4, 5, 6]
df = pd.DataFrame({'study_hours': study_hours, 'marks': marks})
X = df[['study_hours']]
y = df['marks']
model = LinearRegression()
model.fit(X, y)
predicted = model.predict([[7]])
print(f"Predicted marks for 7 study hours: {predicted[0]:.2f}")

# Plot regression line
plt.scatter(study_hours, marks, color='blue', label='Actual')
plt.plot(study_hours + [7], list(model.predict(X)) + [predicted[0]], color='red', label='Regression Line')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.title('Regression: Marks vs Study Hours')
plt.legend()
plt.show()
```

## Classification: Building a Simple Spam Detector

Create a labeled dataset with simple features for spam detection. Train a DecisionTreeClassifier and predict labels for new samples.

```python
# Create simple spam dataset
X = [[0, 1], [1, 0], [0, 0], [1, 1]]  # Features
y = [0, 1, 0, 1]  # Labels (0=not spam, 1=spam)
model = DecisionTreeClassifier()
model.fit(X, y)
sample = [[0, 1]]
predicted = model.predict(sample)
print(f"Predicted label for {sample[0]}: {'Spam' if predicted[0] else 'Not Spam'}")
```

## Clustering: Grouping Points and Visualization

Generate a set of 2D points, apply KMeans clustering, and visualize the clusters using matplotlib.

```python
# Generate 2D points
X = np.array([[2, 60], [3, 65], [4, 70], [5, 75], [6, 80], [7, 85]])
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
labels = kmeans.labels_
print(f"Cluster labels: {labels}")

# Visualize clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', label='Points')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.title('KMeans Clustering: Study Hours vs Marks')
plt.show()
```
